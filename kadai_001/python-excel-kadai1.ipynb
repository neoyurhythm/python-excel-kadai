```python
import openpyxl
from datetime import datetime


#ワークブックの作成
workbook = openpyxl.Workbook()

#アクティブなワークシートを取得する
ws = workbook.active

#固定データの値をセルに設定
ws['B2'] = "請求書"
ws['B4'] = "株式会社ABC"
ws['B5'] = "〒101-0022 東京都千代田区神田練塀町300"
ws['B6'] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws['B7'] = "担当者名:鈴木一郎 様"
ws['F4'] = "№"
ws['F5'] = "日付"
ws['B10'] = "商品名"
ws['C10'] = "数量"
ws['D10'] = "単価"
ws['E10'] = "金額"
ws['B15'] = "小計"
ws['B16'] = "消費税"
ws['B17'] = "合計"
ws['G4'] = "0001"
# G5セルに今日の日付を設定
today = datetime.now().strftime("%Y/%m/%d")
ws['G5'] = today

#請求データの設定
products = {'商品A': [2,10000],
            '商品B': [1,15000]
}

#請求データをセルに書き込む
row = 11  # 商品データの開始行
for product, data in products.items():
    ws.cell(row=row, column=2, value=product)  # 商品名
    ws.cell(row=row, column=3, value=data[0])  # 数量
    ws.cell(row=row, column=4, value=data[1])  # 単価
    ws.cell(row=row, column=5, value=data[0] * data[1])  # 合計金額
    row += 1

# 小計、消費税、合計の計算
# E11とE12の値を取得し、合計を計算
value_e11 = ws['E11'].value
value_e12 = ws['E12'].value
ws['E13'] = value_e11 + value_e12    #小計（空白）
ws['E15'] = value_e11 + value_e12    #税抜き合計（小計）

# 消費税の計算
subtotal = ws['E15'].value    #E15(税抜き合計)の値を取得
ws['E16'] = subtotal*0.1

# E16(消費税額の値を取得)
value_e16 = ws['E16'].value

# 税込額を計算
ws['E17'] = subtotal + value_e16


#ファイル名に日付を追加
now = datetime.now()
file_name = f'請求書_{now.strftime("%Y%m%d")}.xlsx'

#ファイルを保存
workbook.save(file_name)
```
